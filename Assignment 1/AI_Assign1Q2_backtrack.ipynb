```python
import sys

'''readFile() function is used to read the file. It takes the path of the file as input from the used. 
It stores the sudoke problem in a 2x2 array.'''
def readFile():
    print("Enter the path of the file")
    p = input()
    f = open(p,"r")
    dict = []
    for line in f:
        list = []
        for x in line.split():
            y = int(x)
            list.append(y)
        dict.append(list)
    return dict

'''This function takes the input:
1) arr: It is the 2x2 array having the final solution

The function prints the array in a proper readable format'''

def print_sudoku(grid_array): 
    for i in range(9): 
        for j in range(9): 
            print(grid_array[i][j], end=" ")
        print ("\n")
  
'''find_empty_location takes as input:
1) arr: It is the 2x2 array
2) temp: the value of temp=[0,0]

The function finds a position in arr having value 0 and stores its position in temp '''
def find_empty_location(grid_array,temp): 
    for row in range(9): 
        for col in range(9): 
            if(grid_array[row][col]==0): 
                temp[0]=row 
                temp[1]=col
                #print(temp)
                return True
    return False
  
'''Returns a boolean which indicates whether any assigned entry in the specified row matches the given number.''' 
def check_row(grid_array,row,num): 
    for i in range(9): 
        if(grid_array[row][i] == num): 
            return False
    return True
  
'''Returns a boolean which indicates whether any assigned entry in the specified column matches the given number. '''
def check_col(grid_array,col,num): 
    for i in range(9): 
        if(grid_array[i][col] == num): 
            return False
    return True
  
'''Returns a boolean which indicates whether any assigned entry within the specified 3x3 box matches the given number''' 
def check_box(grid_array,row,col,num): 
    for i in range(3): 
        for j in range(3):
            row = row-row%3
            col = col-col%3
            #print("the row checked is",i)
            #print("the col checked is",j)
            if(grid_array[i+row][j+col] == num): 
                return False
    return True
  
''' sudoku_backtracking() takes as input:
1) grid_array: It is a partially solved 2x2 array
2) steps: this is a variable used to count the number of steps taken to reach the solution

The function takes the partially solved array, finds its empty locations, 
checks the constraints on them and assigns values to the array. It returns a fully solved array'''

def solve_sudoku(grid_array): 
    global COUNTER
      
    '''temp is a list variable that keeps the record of row and col in find_empty_location Function     '''
    temp=[0,0] 
      
    ''' If there is no unassigned location, we are done     '''
    if(not find_empty_location(grid_array,temp)):
        print("The steps are:",COUNTER)
        return True
      
    ''' Assigning list values to row and col that we got from the above Function  '''
    row=temp[0] 
    col=temp[1] 
    
      
    ''' consider digits 1 to 9 '''
    for num in range(1,10): 
          
        '''check the constraints '''
        if(check_row(grid_array,row,num)):
            if(check_col(grid_array,col,num)):
                if(check_box(grid_array,row,col,num)):
                 
                    grid_array[row][col]=num
                    
                    COUNTER = COUNTER+1
                    #print("step", COUNTER)
                    if(COUNTER>=10000):
                        sys.exit("steps exceed 10000")

                    '''return, if success '''
                    if(solve_sudoku(grid_array)):
                        return True

                    ''' failure, unmake & try again '''
                    grid_array[row][col] = 0

    ''' this triggers backtracking         '''
    return False 
  
''' Driver main function to test above functions '''
COUNTER = 0

if __name__=="__main__": 
    
    #steps=0
    grid = readFile()
    print(grid)
      
    ''' if success print the grid '''
    if(solve_sudoku(grid)): 
        print_sudoku(grid)
        
    else: 
        print("No solution exists")
```
